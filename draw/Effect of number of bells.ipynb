```python
from matplotlib import pyplot as plt
import numpy as np

X=[1,3,5,10,20]
G_R=[0.0619,0.0711,0.097,0.1126,0.1246]
G_T=[4772,5180,5948,6224,6396]
S_R=[0.0598,0.0631,0.0661,0.0697,0.0751]
S_T=[4656,4824,5056,5132,5340]


fig_R=plt.figure('Effect of number of bells by Requests')
plt.plot(X,G_R, color='#ff00e3', linestyle='-',linewidth=3, marker='^', markerfacecolor='red', markersize=10,label='Greedy')
plt.plot(X,S_R, color='#cadaff', linestyle='-',linewidth=3, marker='o', markerfacecolor='blue', markersize=10,label='Select sort')
plt.legend(loc='upper left')
plt.xlabel('No. of bells')
plt.ylabel('Avg. No. of Ex. Rqts')
#plt.show()
fig_R.savefig('Effect of number of bells by Requests')

fig_T=plt.figure('Effect of number of bells by Time')
plt.plot(X,G_T, color='#ff00e3', linestyle='-',linewidth=3, marker='^', markerfacecolor='red', markersize=10,label='Greedy')
plt.plot(X,S_T, color='#cadaff', linestyle='-',linewidth=3, marker='o', markerfacecolor='blue', markersize=10,label='Select sort')
plt.legend(loc='upper left')
plt.xlabel('No. of bells')
plt.ylabel('Avg. Response Time(ms)')
#plt.show()
fig_T.savefig('Effect of number of bells by Time')
```
