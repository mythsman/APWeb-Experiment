```python
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
X=[400,600,800,1000,1200,1400,1600,1800,2000]
G_R=[0.1639 ,0.1240 ,0.1060 ,0.0972 ,0.0843 ,0.0703 ,0.0697 ,0.0643 ,0.0460]
G_T=[1.485 ,2.391 ,3.387 ,4.458 ,5.463 ,6.627 ,7.947 ,9.198 ,8.739 ]
S_R=[0.1534 ,0.1157 ,0.0871 ,0.0729 ,0.0623 ,0.0527 ,0.0459 ,0.0384 ,0.0376 ]
S_T=[1.422 ,2.304 ,3.078 ,3.942 ,4.791 ,5.556 ,6.333 ,6.891 ,8.007 ]

fig_R=plt.figure('Effect of number of queries by Requests')
plt.plot(X,G_R, color='#ff00e3', linestyle='--',linewidth=3, marker='^', markerfacecolor='red', markersize=10,label='Greedy')
plt.plot(X,S_R, color='#cadaff', linestyle='-',linewidth=3, marker='o', markerfacecolor='blue', markersize=10,label='Select sort')
plt.legend(loc='upper right')
plt.xlabel('No. of Queries')
plt.ylabel('Avg. No. of Ex. Rqts')
fig_R.savefig('Effect of number of queries by Requests')

fig_T=plt.figure('Effect of number of queries by Time')
plt.plot(X,G_T, color='#ff00e3', linestyle='--',linewidth=3, marker='^', markerfacecolor='red', markersize=10,label='Greedy')
plt.plot(X,S_T, color='#cadaff', linestyle='-',linewidth=3, marker='o', markerfacecolor='blue', markersize=10,label='Select sort')
plt.legend(loc='upper left')
plt.xlabel('No. of Queries')
plt.ylabel('Avg. Response Time(ms)')
fig_T.savefig('Effect of number of queries by Time')
```
